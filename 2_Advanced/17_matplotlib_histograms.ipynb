```python
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset=load_dataset('lukebarousse/data_jobs')
df=dataset['train'].to_pandas()

#data cleaning
df['job_posted_date']=pd.to_datetime(df['job_posted_date'])
df['job_skills']=df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
df.head()
```

```python
df_DA_US=df[(df['job_country']=='United States')&(df['job_title']=='Data Analyst')].copy()
```

```python
df_DA_US['salary_year_avg'].plot(kind='hist',bins=30,edgecolor='black')
ax=plt.gca()
plt.xlim(0,250000)
ax.set_title('Data Analyst Salaries in the US')
ax.set_xlabel('Salary')
ax.set_ylabel('no. of jobs')
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: '${:,}K'.format(int(x/1000))))

plt.show()
```
